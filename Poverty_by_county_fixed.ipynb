```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
style.use('ggplot')
```

```python
file = 'Under 18 in Poverty (2011-2016)'
poverty_df = pd.read_csv(file)
poverty_df
```

```python
poverty_df.dtypes
```

```python
poverty_count = poverty_df['Under Age 18 in Poverty Count'].str.replace(',','').astype(int)
```

```python
poverty_df['Under Age 18 in Poverty Count'] = poverty_count
poverty_df
```

```python
poverty_df.dtypes
```

```python
county_group = poverty_df.groupby(['County Name (State)']).mean()
county_group.head()
```

```python
name = county_group["County Name (State)"]
count = county_group["Under Age 18 in Poverty Count"]

bars = plt.bar(name, count, align='center',edgecolor='black')
plt.show

#plt.savefig("Twitter bar graph.png")
```

```python
#county = poverty_df.set_index(["County Name (State)"])["Year"]

county_group = poverty_df.groupby('County Name (State)')
avg_poverty_count = round(county_group['Under Age 18 in Poverty Count'].mean(), 2)
avg_poverty_percent = round(county_group['Under Age 18 in Poverty Percent'].mean(), 2)

#poverty_percentage = poverty_df.groupby["County Name (State)"]("Under Age 18 in Poverty Percent").mean()

poverty = pd.DataFrame({'Poverty Count Average': avg_poverty_count,
                        'Poverty Percentage Average': avg_poverty_percent})
poverty
```

```python
poverty.to_csv("Average_poverty_data", index=False, header=True)
```

```python
sns.barplot(x = "Poverty Count Average", y = "Poverty Percentage Average", data = poverty)
plt.title("Poverty Average Count By County")
plt.xlabel("County")
plt.ylabel("Poverty Count Average")
plt.show()

#plt.savefig("Twitter bar graph.png")
```
